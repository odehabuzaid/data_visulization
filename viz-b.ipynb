```python
custom_colors = ["#4e89ae", "#c56183","#ed6663","#ffa372"]
```

```python
def triple_plot(x, title,c):
    fig, ax = plt.subplots(3,1,figsize=(25,10),sharex=True)
    sns.distplot(x, ax=ax[0],color=c)
    ax[0].set(xlabel=None)
    ax[0].set_title('Histogram + KDE')
    sns.histplot(x, ax=ax[1],color=c)
    ax[1].set(xlabel=None)
    ax[1].set_title('Histplot')
    sns.violinplot(x, ax=ax[2],color=c)
    ax[2].set(xlabel=None)
    ax[2].set_title('Violin plot')
    fig.suptitle(title, fontsize=30)
    plt.tight_layout(pad=3.0)
    plt.show()
```

```python
triple_plot(df_bitcoin['Open'],'Distribution of Opening price in USD of bitcoin 2013 - 2021',custom_colors[0])
```

```python
labels = list(df_bitcoin['Date'])
last_seen = ['2021']
index = [0]
for i in range(len(labels)):
    labels[i] = labels[i].split('-')[0]
    if labels[i] != last_seen[-1]:
        last_seen.append(labels[i])
        index.append(i)

plt.plot(df_bitcoin['High'], label='Bitcoin High Price')
plt.plot(df_bitcoin['Low'], label='Bitcoin Low Price')
plt.plot(df_bitcoin['Open'], label='Bitcoin Open Price')
plt.plot(df_bitcoin['Marketcap'], label='Bitcoin Market Cap')


plt.plot(df_Dogecoin['High'], label='Dogecoin High Price')
plt.plot(df_Dogecoin['Low'], label='Dogecoin Close Price')
plt.plot(df_Dogecoin['Open'], label='Dogecoin Open Price')
plt.plot(df_Dogecoin['Marketcap'], label='Dogecoin Market Cap')


plt.gca().set_yscale("log")
plt.yticks([])
plt.xticks(index, last_seen)
plt.xlabel('Year', labelpad=20)
plt.gcf().set_size_inches(15, 10)
plt.legend()
plt.suptitle('Bitcoin Vs Doogie coin   Open , High Low comparsion', fontsize=30)
plt.tight_layout(pad=3.0)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
seed=42 # random seed for reproductibility

# split data into train and validation
x = df_bitcoin[['High']].values
y = df_bitcoin.iloc[:, 0].values

x_train, x_val, y_train, y_val = train_test_split(x,y,test_size=0.40,random_state=40)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shape {x_val.shape}")
print(f"y_test shape {y_val.shape}")
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
y_predict = lm.predict(x_val)
print(f"Train accuracy {round(lm.score(x_train,y_train)*100,2)} %")
print(f"Test accuracy {round(lm.score(x_val,y_val)*100,2)} %")
```

```python
plt.plot(df_bitcoin['Close'], label='Closing Price Price (USD)')
plt.plot(lm.predict(x), label='Prediction')
plt.gca().set_yscale("log")
plt.xticks(index, last_seen)
plt.xlabel('Year', labelpad=20)
plt.gcf().set_size_inches(15, 10)
plt.legend()
plt.show()
```
